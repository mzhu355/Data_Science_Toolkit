## Second merge

```python
# Second dataset
songs = pd.read_csv("genres_v2.csv")
songs = songs.rename(columns={'song_name': 'Track Name', 'uri': 'Track URI'})
songs
```

```python
playlists.columns
```

```python
# merged_dat['Album Release Date'].nunique()
# merged_dat['Album Release Date']
```

```python
# merged[merged['Pid'] == 992].shape
```

```python
new_merge = pd.merge(songs, playlists, how='inner')
new_merge.shape
new_merge
```

```python
# very few songs are common between the two merges, 
# so both songs datasets are useful to have additional features of songs
print("first merge", merged.shape)
print("second merge", new_merge.shape)
print("overlap between the two merges", pd.merge(new_merge, merged, how='inner').shape)
```

We are going to stack both datasets so we need them to have the same number of columns, and the same column names. Hence remove the useless columns.

```python
# new_merge[new_merge["Pid"]==992].shape
```

```python
merged.columns
```

```python
new_merge.columns
```

```python
merged2 = merged.drop(['Artist URI(s)', 'Album URI', 'Album Artist URI(s)', 'Album Artist Name(s)', 
             'Album Image URL', 'Disc Number', 'Track Preview URL', 'ISRC', 'Added By', 
            'Added At', 'Track Number', 'Explicit','Popularity', 'Album Genres', 'Label', 'Copyrights'], axis = 1)
merged2.shape
```

```python
new_merge2 = new_merge.rename(columns=lambda x: x.title())
new_merge2 = new_merge2.rename(columns={'Duration_Ms':'Track Duration (ms)', 'Genre': 'Artist Genres', 
                           'Track Uri': 'Track URI', 'Time_Signature': 'Time Signature', 
                                        'Artist Name(S)':'Artist Name(s)'})
new_merge2 = new_merge2.drop(['Analysis_Url', 'Unnamed: 0', 'Title', 'Track_Href', 'Id', 'Type'], axis=1)
new_merge2["Album Release Date"] = np.repeat("NaN", len(new_merge2))
new_merge2.columns
new_merge2 = new_merge2.reindex(columns=['Track URI', 'Track Name', 'Artist Name(s)', 'Album Name',
       'Album Release Date', 'Track Duration (ms)', 'Artist Genres',
       'Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
       'Time Signature', 'Playlist', 'Num_Tracks', 'Num_Albums',
       'Num_Artists', 'Follow', 'Collab', 'Pid'])
new_merge2.shape
```

Note: the second dataset doesn't have the release dates of songs

```python
new_merge2
```

## Concatenating both merges

```python
playlists_feat = pd.concat([merged2, new_merge2], axis=0)
playlists_feat = playlists_feat.sort_values(by='Pid')
playlists_feat = playlists_feat.rename(columns={'Artist Name(s)': 'Artist', 'Track Name': 'Track', 'Album Name': 'Album',
                                    'Album Release Date': 'Date', 'Track Duration (ms)': 'Duration', 
                                    'Artist Genres': 'Genre'})
playlists_feat
```
